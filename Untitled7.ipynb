```python
import os
import glob
import pandas as pd
from sklearn.model_selection import train_test_split

os.makedirs("data/processed", exist_ok=True)

# Search broadly inside data/raw/ for any .wav files
files = glob.glob("data/raw/**/*.wav", recursive=True)
print(f"Found {len(files)} audio files.")

if len(files) == 0:
    print("ERROR: Files are missing. Kaggle download might have failed.")
else:
    data = []
    for f in files:
        filename = os.path.basename(f)
        parts = filename.replace('.wav', '').split('_')
        label = parts[-1].lower()
        target_word = parts[1].lower() if len(parts) >= 3 else "youth"
        data.append({"path": f, "text": f"say the word {target_word}", "label": label})

    df = pd.DataFrame(data)
    train_df, test_df = train_test_split(df, test_size=0.3, random_state=42, stratify=df['label'])
    train_df['split'] = 'train'
    test_df['split'] = 'test'

    pd.concat([train_df, test_df]).to_csv("data/processed/manifest.csv", index=False)
    print("Manifest successfully built! Ready for training.")
```
